```python
import pandas as pd
import pyconll
import json
```

```python
conllu_file = './en_ewt-ud-dev.conllu'
data = pyconll.load_from_file(conllu_file)

print(type(data))
print(len(data))
# print(data[0].text)

outfile = 'en_ewt-ud-dev-pos.json'
c = 0
with open(outfile, 'w', encoding='utf-8') as f:
    for sentence in data:
        if len(sentence.text.split()) > 1:
            try:
                item = {
                    'sentence': sentence.text,
                    'tokens': ' '.join([token.form for token in sentence]),
                    'upos': ' '.join([token.upos for token in sentence]),
                    'xpos': ' '.join([token.xpos for token in sentence]),
                    }
            
                f.write(f'{json.dumps(item, ensure_ascii=False)}\n')
                c += 1
            except Exception as e:
                print(e, sentence.text)

print(c)
```
